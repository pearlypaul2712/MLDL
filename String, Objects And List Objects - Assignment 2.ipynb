###  Create the below pattern using nested for loop in Python.

*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*



```python
v = int(input("Enter the number of rows"))
for i in range(0,v):
    for j in range(i+1):
        print("*",end=" ")
    print(" ")
for x in range(v,0,-1):
    for v in range(x-1):
        print("*",end=' ')
    print(" ")
        
```

### Write a Python program to reverse a word after accepting the input from the user.

```python
word = input("Enter the word to be reversed")

print("The reversed word is : {}" .format(word[::-1]))
```
